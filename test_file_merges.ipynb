```python
""" 

Code to explore the feasbility of merging MTA turnstile data with MTA station information file.
Determined through checks here this is not possible.

Searched for lookup file to make merge possible and found: 
https://qri.cloud/nyc-transit-data/remote_complex_lookup

After further examination in code cells below, determined would have to be interested in stations 
at the 'complex' level in order to join the turnstile data with the station information file. 
One complex is multiple stations. 

"""
```

```python
import pandas as pd
import pickle
```

```python
# 2018 turnstile RAW in pickle 'mta_2018'

with open('mta_2018', 'rb') as read_file:
    mta_2018_df = pickle.load(read_file)
    
```

```python
mta_2018_df.head(2) 
```

```python
# Unit is "Remote Unit for a station"
mta_2018_df['UNIT'].unique()
```

```python
# the csv with station info plus added the zip codes
station_info = pd.read_csv('station_info_zipcode.csv')
station_info.head(2)
```

```python
station_info['GTFS Stop ID'].unique()
```

```python
mta_2018_df[mta_2018_df['UNIT'] == 'R01']

# R01 does not exist in turnstile data
# checked turnstile data for several other 'GTFS Stop ID' in station info file

# Though the 'GTFS Stop ID' looks similar to turnstile 'remote unit'  
# these are not the same and cannot merge on this field
```

```python
station_info['Complex ID'].unique()
```

```python
station_info['Station ID'].unique()
```

```python
# the lookup csv file from qri page
lookup = pd.read_csv('MTA_RemoteUnit_Complex_Lookup.csv')
lookup.head(2)

```

```python
# the MTA Remote-Booth-Station.xls

rb_station = pd.read_excel('Remote-Booth-Station.xls')
rb_station.head()
```

```python
## add COMPLEX ID to the station info df
merge_rbs = pd.merge(rb_station, lookup, how='left', left_on=['Remote','Booth'], right_on=['remote','booth'])
merge_rbs.head(2)
```

```python
merge_rbs['complex_id'] = merge_rbs['complex_id'].fillna(0)
merge_rbs.head(2)
```

```python
# fix complex_id so it is not a float
merge_rbs['Complex ID'] = merge_rbs['complex_id'].astype(int)
```

```python
# sample from turnstile data
smp_turnstile_raw = mta_2018_df.sample(25)
smp_turnstile_raw
```

```python
# join turnstile and lookup file from qri

merge_one = pd.merge(smp_turnstile_raw, merge_rbs, how='left', left_on=['UNIT', 'C/A' ], right_on=['Remote', 'Booth' ])


merge_one
```

```python
merge_one.isna().sum()
```

```python

## even with small sample, not every station in turnstile data is in qri lookup
```

```python
merge_two = pd.merge(merge_one, station_info, how='left', left_on= 'complex_id', right_on='Complex ID')

merge_two
```
